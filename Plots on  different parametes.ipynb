## Plotting based on Gender

```python
mx=sns.countplot(x='Gender',data=df)

for bars in mx.containers:
    mx.bar_label(bars)
```

```python
m1=df.groupby(['Gender'],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
m2=sns.barplot(x='Gender',y='Amount',data=m1)

for bars in m2.containers:
    m2.bar_label(bars)
```

## Graphing based on age group

```python
m3=df.groupby("Age Group",as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
m4=sns.barplot(x="Age Group",y="Amount",data=m3)
for bars in m4.containers:
    m4.bar_label(bars)
sns.set({"figure.figsize":(15,9)})
```

```python
m6=df.groupby("Age Group",as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False)
m7=sns.barplot(x="Age Group",y="Orders",data=m6)
for bars in m7.containers:
    m7.bar_label(bars)
```

```python
m7=sns.countplot(x="Age Group",data=df,hue="Gender")
for bars in m7.containers:
    m7.bar_label(bars)
```

```python
m8=sns.countplot(x="Age Group",data=df)
```

## Plots based on States

```python
m7=df.groupby("State",as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
m8=sns.barplot(data=m7,x="State",y="Amount")
for bars in m8.containers:
    m8.bar_label(bars)
sns.set(rc={'figure.figsize':(25,5)})
```
